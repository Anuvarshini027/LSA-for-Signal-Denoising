```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
signal=np.loadtxt("ecg.txt")
y1=signal[:,0]
y2=((y1-max(y1)))/(max(y1)-min(y1)) #normalising to bring the values to a same range, here from [-1,0]
noisy_signal=y2+ np.random.randn(len(y2)) * 0.01 #noise added
noisy_signal.shape
```

```python
def leastSquareDenoising(noisy_signal):
    """Generates the denoised signal from the noisy signal
    Args:
        noisy signal => signal with noise added
    """
    N=len(noisy_signal)
    d_mat=np.eye(N-2,N) #Identity matrix of size (N-2,N)
    for i in range(0,N-2):
        d_mat[i,i+1]=-2 #1,-2,1 #SECOND ORDER DIFFERENTIAL MATRIX
        d_mat[i,i+2]=1
    I=np.eye(N,N) #identity matrix    
    for lam in [1,10,100,1000,10000,100000,500000]:
        denoised=I+lam*(np.dot(np.transpose(d_mat),d_mat)) 
        #(I+lam*(d_mat^T*d_mat))*signal,denoised=NxN+NxN=NxN
        denoise_inv=np.linalg.inv(denoised)  #inverse
        denoised_signal=np.dot(denoise_inv,noisy_signal) #(NxN)*(Nx1)=Nx1       
        plt.plot(denoised_signal)
        plt.title(f"Denoised Signal when Lamda = {lam} ")
        plt.show()       
    return None
```

```python
plt.plot(noisy_signal)
plt.title("Signal with Noise")
plt.show()
leastSquareDenoising(noisy_signal)
```
